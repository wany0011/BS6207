```python
import torch
import torch.nn.functional as F
import sys
import os
 
torch.manual_seed(2020)
 
# Number of hidden layers
K=10
# number of input nodes
d=10
# N is batch size; D_in is input dimension;
# H is hidden dimension; D_out is output dimension.
N, D_in, H, D_out = 1, d, [2*d+1,K], 1

 
# option 1 : All weights starting from Zero and ALL bias from One.

w0=torch.zeros((D_in, H[0]), dtype=torch.float,requires_grad=True)
b0=torch.ones(H[0],dtype=torch.float,requires_grad=True)

w1=torch.zeros((H[0], H[0]), dtype=torch.float,requires_grad=True)
b1=torch.ones(H[0],dtype=torch.float,requires_grad=True)

w2=torch.zeros((H[0], H[0]), dtype=torch.float,requires_grad=True)
b2=torch.ones(H[0],dtype=torch.float,requires_grad=True)


w3=torch.zeros((H[0], H[0]), dtype=torch.float,requires_grad=True)
b3=torch.ones(H[0],dtype=torch.float,requires_grad=True)

w4=torch.zeros((H[0], H[0]), dtype=torch.float,requires_grad=True)
b4=torch.ones(H[0],dtype=torch.float,requires_grad=True)

w5=torch.zeros((H[0], H[0]), dtype=torch.float,requires_grad=True)
b5=torch.ones(H[0],dtype=torch.float,requires_grad=True)

w6=torch.zeros((H[0], H[0]), dtype=torch.float,requires_grad=True)
b6=torch.ones(H[0],dtype=torch.float,requires_grad=True)

w6=torch.zeros((H[0], H[0]), dtype=torch.float,requires_grad=True)
b6=torch.ones(H[0],dtype=torch.float,requires_grad=True)

w7=torch.zeros((H[0], H[0]), dtype=torch.float,requires_grad=True)
b7=torch.ones(H[0],dtype=torch.float,requires_grad=True)

w8=torch.zeros((H[0], H[0]), dtype=torch.float,requires_grad=True)
b8=torch.ones(H[0],dtype=torch.float,requires_grad=True)

w9=torch.zeros((H[0], D_out), dtype=torch.float,requires_grad=True)
b9=torch.ones(D_out,dtype=torch.float,requires_grad=True)
"""
# option 2 : randomized weights and bias over [0,1] at starting point 

w0=torch.rand((D_in, H[0]), dtype=torch.float,requires_grad=True)
b0=torch.rand(H[0],dtype=torch.float,requires_grad=True)

w1=torch.rand((H[0], H[0]), dtype=torch.float,requires_grad=True)
b1=torch.rand(H[0],dtype=torch.float,requires_grad=True)

w2=torch.rand((H[0], H[0]), dtype=torch.float,requires_grad=True)
b2=torch.rand(H[0],dtype=torch.float,requires_grad=True)


w3=torch.rand((H[0], H[0]), dtype=torch.float,requires_grad=True)
b3=torch.rand(H[0],dtype=torch.float,requires_grad=True)

w4=torch.rand((H[0], H[0]), dtype=torch.float,requires_grad=True)
b4=torch.rand(H[0],dtype=torch.float,requires_grad=True)

w5=torch.rand((H[0], H[0]), dtype=torch.float,requires_grad=True)
b5=torch.rand(H[0],dtype=torch.float,requires_grad=True)

w6=torch.rand((H[0], H[0]), dtype=torch.float,requires_grad=True)
b6=torch.rand(H[0],dtype=torch.float,requires_grad=True)

w6=torch.rand((H[0], H[0]), dtype=torch.float,requires_grad=True)
b6=torch.rand(H[0],dtype=torch.float,requires_grad=True)

w7=torch.rand((H[0], H[0]), dtype=torch.float,requires_grad=True)
b7=torch.rand(H[0],dtype=torch.float,requires_grad=True)

w8=torch.rand((H[0], H[0]), dtype=torch.float,requires_grad=True)
b8=torch.rand(H[0],dtype=torch.float,requires_grad=True)

w9=torch.rand((H[0], D_out), dtype=torch.float,requires_grad=True)
b9=torch.rand(D_out,dtype=torch.float,requires_grad=True)

"""

cost_list=[] 
for e in range(epoches):
    y0 = F.relu(torch.mm(x_train,w0)+b0)
    y1 = F.relu(torch.mm(y0,w1)+b1)
    y2 = F.relu(torch.mm(y1,w2)+b2)
    y3 = F.relu(torch.mm(y2,w3)+b3)
    y4 = F.relu(torch.mm(y3,w4)+b4)
    y5 = F.relu(torch.mm(y4,w5)+b5)
    y6 = F.relu(torch.mm(y5,w6)+b6)
    y7 = F.relu(torch.mm(y6,w7)+b7)
    y8 = F.relu(torch.mm(y7,w8)+b8)
    y9 = F.relu(torch.mm(y8,w9)+b9)
    
    
    #print("y0={}".format(list(y0)))
    #print("y1={}".format(list(y1)))
    #print("y2={}".format(list(y2)))
    #print("y3={}".format(list(y3)))
    #print("y4={}".format(list(y4)))
    #print("y5={}".format(list(y5)))
    #print("y6={}".format(list(y6)))
    #print("y7={}".format(list(y7)))
    #print("y8={}".format(list(y8)))
    #print("y9={}".format(list(y9)))
    
    
    lost = torch.mean((y9 - y_train) ** 2)
    cost_list.append(lost.item())
    
    lost.backward()
    
    if True:# e==0:
        file_name='torch_autograd_epoch'+str(e)+'.dat'
        
        if os.path.exists(file_name):
            os.remove(file_name) #this deletes the file
            
        original_stdout = sys.stdout # Save a reference to the original standard output
        
        with open(file_name, 'w') as f:
            sys.stdout = f # Change the standard output to the file we created.
            print("[layer 0]")
            print(w0.grad.data)
            print(b0.grad.data)
    
            print("[layer 1]")
            print(w1.grad.data)
            print(b1.grad.data)
    
            print("layer 2")
            print(w2.grad.data)
            print(b2.grad.data)

            print("layer 3")
            print(w3.grad.data)
            print(b3.grad.data)

            print("layer 4")
            print(w4.grad.data)
            print(b4.grad.data)

            print("[layer 5]")
            print(w5.grad.data)
            print(b5.grad.data)
    
            print("[layer 6]")
            print(w6.grad.data)
            print(b6.grad.data)
    
            print("layer 7")
            print(w7.grad.data)
            print(b7.grad.data)

            print("layer 8")
            print(w8.grad.data)
            print(b8.grad.data)

            print("layer 9")
            print(w9.grad.data)
            print(b9.grad.data)

            
        sys.stdout = original_stdout # Reset the standard output to its original value    
        #print(w0.grad.data)
        #print(b0.grad.data)
    
    # Update Parameters
    w0.data = w0.data - learning_rate * w0.grad.data
    b0.data = b0.data - learning_rate * b0.grad.data
    
    w1.data = w1.data - learning_rate * w1.grad.data
    b1.data = b1.data - learning_rate * b1.grad.data
    
    w2.data = w2.data - learning_rate * w2.grad.data
    b2.data = b2.data - learning_rate * b2.grad.data
    
    w3.data = w3.data - learning_rate * w3.grad.data
    b3.data = b3.data - learning_rate * b3.grad.data
    
    w4.data = w4.data - learning_rate * w4.grad.data
    b4.data = b4.data - learning_rate * b4.grad.data
    
    w5.data = w5.data - learning_rate * w5.grad.data
    b5.data = b5.data - learning_rate * b5.grad.data
    
    w6.data = w6.data - learning_rate * w6.grad.data
    b6.data = b6.data - learning_rate * b6.grad.data
    
    w7.data = w7.data - learning_rate * w7.grad.data
    b7.data = b7.data - learning_rate * b7.grad.data
    
    w8.data = w8.data - learning_rate * w8.grad.data
    b8.data = b8.data - learning_rate * b8.grad.data
    
    w9.data = w9.data - learning_rate * w9.grad.data
    b9.data = b9.data - learning_rate * b9.grad.data
        
    
    print('epoch {}, Loss: {:.5f}'.format(e, cost_list[e]))
    w0.grad.data.zero_()
    b0.grad.data.zero_()
    
    w1.grad.data.zero_()
    b1.grad.data.zero_()
    
    w2.grad.data.zero_()
    b2.grad.data.zero_()

    w3.grad.data.zero_()
    b3.grad.data.zero_()
        
    w4.grad.data.zero_()
    b4.grad.data.zero_()

    w4.grad.data.zero_()
    b4.grad.data.zero_()

    w5.grad.data.zero_()
    b5.grad.data.zero_()

    w6.grad.data.zero_()
    b6.grad.data.zero_()

    w7.grad.data.zero_()
    b7.grad.data.zero_()

    w8.grad.data.zero_()
    b8.grad.data.zero_()

    w9.grad.data.zero_()
    b9.grad.data.zero_()
```
